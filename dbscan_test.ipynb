## Exemple tirée d'une implémentation R

```python
import pyreadr

result = pyreadr.read_r('/Users/leomonnier/Documents/M2DS/Projet_TER/multishapes.rda') # also works for Rds

# done! let's see what we got
# result is a dictionary where keys are the name of objects and the values python
# objects
print(result.keys()) # let's check what objects we got
df1 = result['multishapes'] # extract the pandas data frame for object df1
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df1['x'],df1['y'])
```

```python
import seaborn as sns
```

```python
from sklearn.cluster import DBSCAN
```

```python
labels=DBSCAN(eps = 0.15, min_samples=5).fit_predict(df1[['x','y']])
```

```python
sns.scatterplot(x=df1['x'],y=df1['y'],hue=labels,palette="Paired_r",style=labels)
```

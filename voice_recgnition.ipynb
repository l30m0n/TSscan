```python
from scipy.io import wavfile
from matplotlib import pyplot as plt
import numpy as np

from scipy.spatial.distance import euclidean
from fastdtw import fastdtw

# Read stored audio files for comparison
fs1, data1 = wavfile.read("./voice/clip1.wav")
fs2, data2 = wavfile.read("./voice/clip2.wav")
fs3, data3 = wavfile.read("./voice/clip3.wav")
fs4, data4 = wavfile.read("./voice/clip4.wav")

# Take the max values along axis
data1 = np.amax(data1, axis=1)
data2 = np.amax(data2, axis=1)
data3 = np.amax(data3, axis=1)
data4 = np.amax(data4, axis=1)

# Create subplots
ax = plt.subplot(2, 2, 1)
ax.plot(data1, color='#67A0DA')
ax.plot(data2,color='red')
ax.plot(data3,color='green')
ax.plot(data4,color='purple')






# Display created figure
fig=plt.show()
display(fig)
```

```python
data=np.array([data1.tolist(),data2.tolist(),data3.tolist(),data4.tolist()])
data[0]
```

```python
tsc=TSscan(eps=100,dist_rescale=False)

tsc.fit(data)
```

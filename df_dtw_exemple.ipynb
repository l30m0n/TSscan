```python
import dtw
import numpy as np
```

Ici on réalise une très simple utilisation des packages de dynamic time warping

```python
data=np.linspace(-1,1)
X=data**2
Y=data*0
```

```python
a=dtw.dtw(X,Y,keep_internals=True)
```

```python
a.plot("twoway")
```

```python
a.distance
```

On voit bien ici que le résultat n'est pas celui de notre définition continue du dynmic time warping.
